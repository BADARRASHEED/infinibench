## 2. Preview a Few Streaming Samples

This reads only a few examples from HuggingFace streaming mode. It does not download the full dataset.

```python
from datasets import load_dataset
from IPython.display import display

dataset = load_dataset("Haoming645/infinibench", split="train", streaming=True)

for i, sample in enumerate(dataset.take(3), start=1):
    print(f"Sample {i}")
    print("Available fields:", list(sample.keys()))
    if "label" in sample:
        print("Label:", sample["label"])
    display(sample["image"])
```

## 6. Fill Answers Manually

Open `data/results_template.csv` and fill these columns:

- `ground_truth`: the correct answer you decide by looking at the image
- `model_answer`: the answer from a model or from manual testing
- `correct`: use `1` if the model answer is correct, or `0` if it is wrong

This Lite project does not call a VLM API by default. You can add model answers later from any source.
